## Retrieve, load and process data

```python
# Import pandas library
import pandas as pd
print('pandas library imported.')
```

### Free or reduced-price lunches per LA school (percent eligibility)

```python
# Read xlsx file containing free lunch program data
url = 'https://www.cde.ca.gov/ds/sd/sd/documents/frpm1819.xlsx'
frLunch_CA = pd.read_excel(url, sheet_name=1, header=1,
                       usecols = ['County Name', 'County Code', 'District Code', 
                                 'School Code', 
                                 'Percent (%) \nEligible FRPM \n(K-12)'])
frLunch_CA.head()
```

```python
#Rename columns
frLunch_CA.rename(columns={'County Name': 'CountyName', 
                        'County Code':'CountyCode', 
                       'District Code':'DistrictCode', 
                       'School Code': 'SchoolCode', 
                       'Percent (%) \nEligible FRPM \n(K-12)': 'PctEligibleFRPM'},
               inplace=True)
frLunch_CA.head()
```

```python
# Extract LA county schools
frLunch_LA = frLunch_CA[frLunch_CA['CountyName'].str.contains('Los Angeles', regex=False)]
frLunch_LA.head()
```

```python
# Check that County Code is unique
pd.unique(frLunch_LA.CountyCode)
```

```python
# Combine codes into one code: CDSCode
CDSCode = []
for n in range(len(frLunch_LA)):
    code = str(str(frLunch_LA.CountyCode.iloc[n]) + 
                  str(frLunch_LA.DistrictCode.iloc[n]) + 
                  str(frLunch_LA.SchoolCode.iloc[n]))
    CDSCode.append(code)
CDSCode
```

```python
# Add list to dataframe
frLunch_LA['CDSCode'] = CDSCode
frLunch_LA.head()
```

```python
# Drop redundant columns
frLunch_LA = frLunch_LA.drop(columns=['CountyCode','DistrictCode','SchoolCode',
                        'CountyName'])
frLunch_LA.head()
```

### LA county school directory

```python
# Read xlsx file containing school district information
url = 'https://www.cde.ca.gov/SchoolDirectory/report?rid=dl1&tp=xlsx&ict=Y'
schoolDir = pd.read_excel(url, header=5, 
                         usecols = ['CDSCode', 'Zip'])
schoolDir.head()
```

```python
# Merge frLunch_LAzip and schoolDir dataframes
frLunch_LAzip = pd.merge(frLunch_LA, schoolDir, on='CDSCode')
frLunch_LAzip
```

```python
# Check for missing Zip codes
frLunch_LAzip[frLunch_LAzip.Zip.isna()]
```

```python
# Edit postal codes to first 5 digits only
nSchools = len(frLunch_LAzip)
for n in range(0,nSchools):
    frLunch_LAzip.Zip.iloc[n] = frLunch_LAzip.iloc[n].Zip[0:5]
frLunch_LAzip
```

```python
# Remove duplicates
frLunch_LAzip.drop_duplicates('CDSCode')
frLunch_LAzip
```

## Retrieve latitude and longitude coordinates of each school

```python
# Import pgeocode library
! pip install pgeocode
import pgeocode
print('pgeocode library installed')
```

```python
# Define country
nomi = pgeocode.Nominatim('us')
```

```python
# Retrieve lat and long from zip code
geoList = []
nSchools = len(frLunch_LAzip)
for n in range(0,nSchools):
    Zip = frLunch_LAzip.iloc[n].Zip
    Lat = nomi.query_postal_code(Zip).latitude
    Lon = nomi.query_postal_code(Zip).longitude
    geoList.append([Zip, Lat, Lon])
geoData = pd.DataFrame(geoList, columns=['Zip', 'Lat', 'Lon'])
geoData = geoData.drop_duplicates('Zip')
geoData
```

```python
# Merge location data with frLunchLAZip dataframe
frLunch_LAgeo = pd.merge(frLunch_LAzip, geoData, on='Zip')
frLunch_LAgeo.sort_values('Lat')
```

## Explore and cluster Los Angeles county schools

As per the NCES, schools can be classified at varying levels of poverty based on the percentage of students whom are eligible for free or reduced-price lunch (FRPL). See https://nces.ed.gov/fastfacts/display.asp?id=898 for the breakdown.

```python
# Focus on schools with high poverty classfication
frLunch_LApov = frLunch_LAgeo[frLunch_LAgeo.PctEligibleFRPM>=0.751]
frLunch_LApov.shape
frLunch_LApov
```

```python
# import Folium library
!conda install -c conda-forge folium=0.5.0 --yes
import folium
#import matplotlib.cm as cm
#import matplotlib.colors as colors
print('folium library imported')
```

```python
# Create LA folium map object
map_LA = folium.Map(location=[34.0522, -118.2437], 
                        zoom_start=10)
```

```python
# Use KMeans clustering to cluster LA county neighbourhoods
# Import KMeans library
from sklearn.cluster import KMeans
print('KMeans library imported')
```

```python
# Import numpy library
import numpy as np
print('numpy library imported')
```

```python
# Prepare dataframe for clustering
coords = frLunch_LApov.drop(columns=['PctEligibleFRPM', 'CDSCode', 'Zip'])
```

```python
coords.sort_values('Lat')
```

```python
# Execute KMeans clustering
k=8
kmeans = KMeans(n_clusters = k).fit(coords)
kmeans.labels_
```

```python
frLunch_LApov.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
frLunch_LApov[frLunch_LApov.notna()]
```

```python
# Visualise neighbourhood clusters on toronto map
# Import numpy and matplotlib libraries
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors
print('numpy and matplotlib libraries imported')
```

```python
# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, cds, cluster in zip(frLunch_LApov['Lat'],
                                            frLunch_LApov['Lon'],
                                            frLunch_LApov['CDSCode'],
                                            frLunch_LApov['Cluster Labels']
                                           ):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_LA)
       
map_LA
```

```python
frLunch_LApov2 = frLunch_LApov.groupby(['Zip', 'Lat','Lon','Cluster Labels']).size().reset_index()
frLunch_LApov2.rename(columns={0:'Count'}, inplace=True)
frLunch_LApov2
```

```python
# Bubble chart to indicate zip codes of high density poverty
map_LA2 = folium.Map(location=[34.0522, -118.2437], 
                        zoom_start=9)
```

```python
for i in range(len(frLunch_LApov2)):
   folium.CircleMarker(
      [frLunch_LApov2.Lat.iloc[i], frLunch_LApov2.Lon.iloc[i]],
      popup=frLunch_LApov2.Zip.iloc[i],
      radius=str(frLunch_LApov2.Count.iloc[i]),
      color=rainbow[frLunch_LApov2.iloc[i]['Cluster Labels']-1],
      fill=True,
      fill_color=rainbow[frLunch_LApov2.iloc[i]['Cluster Labels']-1],
      fill_opacity=0.7).add_to(map_LA2)

```
