# Hungry Kids and the LA County School Shutdown Amid COVID-19 Pandemic

## Introduction/Business Problem

In efforts to control the spread of the new highly infectious coronavirus, LAUSD, the nation's second largest school district has cancelled on-campus attendance indefinitely. Expanding this to the entire Los Angeles county indicates that over 1 million school-aged children who had qualified for free or reduced-price meals at schools may now go hungry.

The question is, what are the areas of most impact of the school closures so that alternative meal distribution sites can be set up to concentrate on those areas. As such, LA county school districts and non-profit organisations would find this information very useful for quick and effective action in meeting the needs of LA's kids facing long term and extreme food poverty.

## Data

Using California's FRPM School-Level Data and school district information, the areas of greatest impact i.e. greatest concentrations of school-age children that may now be experiencing extreme food poverty without access to school-meals. Additionally, the Foursquare location data can be leveraged to find facilities local to the target areas and visualise how pop-up food distribution points may be set up with accessibility in mind, while ahering to the WHO recommendations for limiting the spread of the virus.

The school-level data on student eligibility for free or reduced-price meals was sourced from California Department of Education via https://www.cde.ca.gov/ds/sd/sd/documents/frpm1819.xlsx, last accessed 17 March 2020. This dataset contains the number and percentage of students whom are eligible for the state's free or reduced-price meals program (FRPM) for each school within the state of California. A subset of LA county schools was pulled from this dataset for this project. The school district information was also sourced from California Department of Education via https://www.cde.ca.gov/SchoolDirectory/report?rid=dl1&tp=xlsx&ict=Y, last accessed 17 March 2020, which contains the addresses of each school in the state. The Foursquare API was utilised to locate nearby facilities within the identified areas of greatest impact of LA county school closures with regard to food instability/poverty.

## Methodology

As previously mentioned, a subset of the California FRPM School-Level dataset which contained only Los Angeles County schools was created. Percent eligible students was used as the target variable. The county, district and school codes were combined into one CDS code for later merging with the school district information dataset which contains the relevant school zipcodes. These postal (zip) codes were reduced to the standard 5-digit format to prescribe to the location data format used by the Foursquare API. The dataset was checked for missing entries and duplicates - of which none were identified. 

This dataset included the percent of student eligible for free or reduced-price meals, the CDS code and postal code for each of the 1686 schools within LA county. 

The pgeocode library was used to retrieve the latitude and longitude coordinates for each school based on its respective postal code, which were stored in a dataframe and later merged with the aforementioned dataset. One missing entry was identified and later list-wise deleted as it was found to be an incorrect zipcode which could not be rectified in a timely manner.

As per the NCES, schools can be classified at varying levels of poverty based on the percentage of students whom are eligible for free or reduced-price lunch (FRPL). See https://nces.ed.gov/fastfacts/display.asp?id=898 for the breakdown. Schools with percent FRPM eligibility above 75% are considered to be high poverty schools. There were 958 of such LA county schools pulled from the dataset. 

These schools were classified into one of 8 clusters using the KMeans library from the sklearn.cluster package. The clustering locations of these schools was then visualised on a folium map using circle marker. A frequency analysis was then conducted to determine the number of schools meeting the mid-high to high poverty classification within each postal code. A new folium map was created using varying marker sizes to indicate concentrations of these poverty schools for easier identification of areas of greatest impact of school shutdown, of the 221 zip codes identifed.

Finally, the requests, geopy.geocoders Nominatim and pandas.io.json json_normalize libraries were used to query nearby parks/playgrounds and parking lots of the zipcodes of concern and pull the respective GPS coordinates. The locations of these facilities were added to the folium map for better visualisation of how pop-up meal distribution points may be coordinated for best effect.

## Discussion & Conclusion

In addition to the temporary meal distribution centers set up in LA county schools, the findings presented in this report revealed the following:
- areas of greatest concern, as illustrated in Figure 1
- additional sites for safe meal distribution (i.e. open space) to improve accessibility in those areas, presented in Figure 2.


The following actions are recommended for most impactful response based on the identified areas of greatest need due to food insecurity. Additional meal distribution sites should be set up:
- Darwell Park in 90201
- Baldwin Park in 91706
- Ralph Welch Park in 91766
- Central Jazz Park in 90011
- Wilson Park in 90221


These recommendations have been provided on the basis of the highest impact at the lowest cost without incurring hosting/site fees that may otherwise be charged by local food establishments. However, it should be noted that there is a distinct lack of parks/playgrounds in areas of high childhood poverty density. Most of what has been located are parking facilities. Thus more effort should be exerted in these areas to ensure the health and wellbeing of these children while schools remain closed.
