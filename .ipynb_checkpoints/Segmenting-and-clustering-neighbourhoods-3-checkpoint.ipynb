## Scrape data from Wikipedia page

```python
# Import pandas library
import pandas as pd
print('pandas library imported.')
```

```python
# Import BeautifulSoup library
from bs4 import BeautifulSoup
import lxml
print('BeautifulSoup and lxml libraries imported.')
```

```python
# Import requests library
import requests
print('requests library imported.')
```

```python
# Download data via URL
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
source = requests.get(url).text
source_data = BeautifulSoup(source,'lxml')
print('data downloaded from ', url)
```

```python
# Create dataframe to store downloaded data with specified column names
colNames = ['PostalCode','Borough', 'Neighborhood']
toronto_df = pd.DataFrame(columns=colNames)
toronto_df.head()
```

```python
# Parse postcode data from Wikipedia page data
CA_data = source_data.find('div', class_='mw-parser-output')
dataTable = str(CA_data.table)
from IPython.display import display_html
display_html(dataTable,raw=True)
```

```python
# Convert html table to pandas dataframe
toronto_df = pd.read_html(dataTable)
toronto_df = toronto_df[0] #return dataframe
toronto_df.head()
```

## Data preprocessing and cleaning

```python
# Ignore cells with a borough that is "Not assigned"
toronto_df[toronto_df.Borough=='Not assigned']
```

```python
Tdf_cleaned = toronto_df[toronto_df.Borough!='Not assigned']
Tdf_cleaned[Tdf_cleaned.Borough=="Not assigned"]
```

```python
# Combine neighbourhoods with the same postcode.
Tdf_cleaned = Tdf_cleaned.groupby(['Postcode', 'Borough'],
                                 sort = False).agg(', '.join)
Tdf_cleaned.reset_index(inplace=True)
Tdf_cleaned.head()
```

```python
# Correct any "Not assigned" neighbourhoods to their respective Boroughs
for n in range(0,len(Tdf_cleaned)):
    if Tdf_cleaned.Neighbourhood[n] == "Not assigned":
        Tdf_cleaned.Neighbourhood[n] = Tdf_cleaned.Borough[n]
Tdf_cleaned[Tdf_cleaned.Neighbourhood=="Not assigned"]
```

```python
Tdf_cleaned.head()
```

```python
# Print the number of rows of dataframe
Tdf_cleaned.shape
       
```

## Retrieve latitude and longitude coordinates of each neighbourhood

```python
# Read csv file containing location data
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

```python
# Merge location data with Toronto dataframe
lat_lon.rename(columns={'Postal Code':'Postcode'}, inplace=True)
Tdf_cleaned = pd.merge(Tdf_cleaned, lat_lon, on='Postcode')
Tdf_cleaned.head(12)
```

## Explore and cluster Toronto neighbourhoods

```python
# Retrieve all rows of Toronto boroughs
toronto = Tdf_cleaned[Tdf_cleaned['Borough'].str.contains('Toronto', regex=False)]
toronto.reset_index(inplace=True)
toronto
```

```python
# import Folium library
!conda install -c conda-forge folium=0.5.0 --yes
import folium
#import matplotlib.cm as cm
#import matplotlib.colors as colors
print('folium library imported')
```

```python
# Visualise Toronto neighbourhoods
map_toronto = folium.Map(location=[43.700110, -79.416300], 
                        zoom_start=11)
for lat, lng, borough, neighbourhood in zip(toronto['Latitude'],
                                            toronto['Longitude'],
                                            toronto['Borough'],
                                            toronto['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker([lat,lng], radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

```python
# Use KMeans clustering to cluster Toronto neighbourhoods
# Import KMeans library
from sklearn.cluster import KMeans
print('KMeans library imported')
```

```python
k=5
toronto_clustering = toronto.drop(['Postcode','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
toronto.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
# Visualise neighbourhood clusters on toronto map
# Import numpy and matplotlib libraries
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors
print('numpy and matplotlib libraries imported')
```

```python
# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(toronto['Latitude'],
                                            toronto['Longitude'],
                                            toronto['Neighbourhood'],
                                            toronto['Cluster Labels']
                                           ):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_toronto)
       
map_toronto
```
